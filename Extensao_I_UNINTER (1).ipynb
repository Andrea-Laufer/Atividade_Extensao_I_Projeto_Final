### VIII - ANÁLISE DAS IES

Neste tópico, objetiva-se:

* quantificar o total de IES cadastradas no e-MEC;

* quantificar quantas estão ativas e quantas estão desativadas no e-MEC;

* quantificar quantas ativas são privadas (nas mais diversas formas de apresentação, como Associações, Fundações, Privadas com Fins Lucrativos ou Sem Fins Lucrativos) e quantas são públicas, isto é, por "categoria administrativa".

Para simplificar, consideramos privada toda natureza jurídica diferente de pública. Assim, associações civis, fundações privadas, comunitárias, confessionais, filantrópicas, todas estão sendo consideradas no conjunto privado de oferta de ensino. Essa abordagem está desta forma, pois, ainda que haja natureza jurídica em IES privada diferente de "fins lucrativos", o fato é que operam em mercado (de ensino) e há aferição de ganhos, mesmo que não distribuídos a dirigentes ou afins.

```python
total_ies = len(df_ies)

# Filtra as ativas e extintas
active_ies = df_ies[df_ies['SITUACAO_IES'] == 'Ativa'].shape[0]
inactive_ies = df_ies[df_ies['SITUACAO_IES'] == 'Extinta'].shape[0]

print(f"Total de IES cadastradas: {total_ies}")
print(f"IES ativas: {active_ies}")
print(f"IES inativas (extintas): {inactive_ies}")
```

a) IES totais, IES ativas x IES inativas:

Do dado acima, vê-se que há um total de 4569 IES cadastradas no e-MEC (sistema de gerenciamento de dados da educação superior do Ministério da Educação) e destas 3040 são ativas, totalizando 66% aproximadamente.

b) Visualização da Distribuição de IES Ativas x Inativas

Para entender melhor a proporção de instituições de ensino superior ativas e inativas, exibe-se um gráfico de barras com base nos dados que já foram processados.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cria um DataFrame para a visualização
# Puxa os números calculados (active_ies e inactive_ies)
data = {
    'Situação': ['Ativa', 'Extinta'],
    'Número de IES': [active_ies, inactive_ies]
}
df_status = pd.DataFrame(data)

# Cria o gráfico de barras
plt.figure(figsize=(8, 6))

# hue='Situação' e legend=False
# Isso é necessário nas versões novas do Seaborn para o gráfico colorir corretamente sem dar erro
sns.barplot(x='Situação', y='Número de IES', data=df_status, palette='viridis', hue='Situação', legend=False)

plt.title('Distribuição de IES Ativas vs. Inativas')
plt.xlabel('Situação da IES')
plt.ylabel('Número de Instituições')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

c) Análise de IES Ativas por Categoria Administrativa

Vamos quantificar as instituições de ensino superior ativas, classificando-as em categorias públicas e privadas.

A análise revela que há uma massiva quantidade de IES privadas em comparação à quantidade de IES públicas.

Em um cálculo manual simples, as IES privadas representam 87% da atuação na oferta do ensino superior do país.

```python
# Filtrar apenas as IES ativas
df_active_institutions = df_ies[df_ies['SITUACAO_IES'] == 'Ativa'].copy()

# Cria uma nova coluna para categorização simplificada (Pública/Privada)
def categorize_ies(categoria):
    # Adicionei str() por segurança, caso haja algum valor vazio na coluna
    if 'Pública' in str(categoria):
        return 'Pública'
    else:
        return 'Privada'

df_active_institutions['TIPO_ADMINISTRATIVO'] = df_active_institutions['CATEGORIA_DA_IES'].apply(categorize_ies)

# Contar a distribuição de IES por tipo administrativo
distribution_by_category = df_active_institutions['TIPO_ADMINISTRATIVO'].value_counts()

print("Distribuição de IES Ativas por Categoria Administrativa (Pública/Privada):")
display(distribution_by_category.to_frame(name='Número de IES'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cria um DataFrame a partir da Series distribution_by_category

df_admin_category = distribution_by_category.reset_index()
df_admin_category.columns = ['Tipo Administrativo', 'Número de IES']

# Cria o gráfico de barras
plt.figure(figsize=(8, 6))

# hue='Tipo Administrativo' e legend=False
# Isso evita o erro de depreciação do Seaborn e garante as cores certas
sns.barplot(
    x='Tipo Administrativo',
    y='Número de IES',
    data=df_admin_category,
    palette='coolwarm',
    hue='Tipo Administrativo',
    legend=False
)

plt.title('Distribuição de IES Ativas por Categoria Administrativa')
plt.xlabel('Tipo Administrativo')
plt.ylabel('Número de Instituições')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

d) Análise de IES Privadas por Unidade Federativa (UF)

Vamos quantificar o número de IES privadas ativas em cada estado (UF).

```python
# Filtra apenas as IES Privadas usando o dataframe que já preparamos
private_ies_by_uf = df_active_institutions[df_active_institutions['TIPO_ADMINISTRATIVO'] == 'Privada']

# Conta quantas existem por Estado (UF)
private_ies_by_uf_count = private_ies_by_uf['UF'].value_counts().sort_values(ascending=False)

print("Número de IES Privadas Ativas por UF:")
display(private_ies_by_uf_count.to_frame(name='Número de IES Privadas'))
```

e) Visualização da Distribuição de IES Privadas por UF

Para ter uma visão clara da concentração de IES privadas em cada estado, exibimos um gráfico de barras horizontal.

A maior concentração aponta para o Estado de SP, seguido por MG, PR, BA. RJ representa a 8a posição.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cria um DataFrame a partir da Series (igual ao original)
df_private_ies_uf = private_ies_by_uf_count.reset_index()
df_private_ies_uf.columns = ['UF', 'Número de IES Privadas']

plt.figure(figsize=(12, 8))

# hue='UF' e legend=False
# Como as barras são por UF, o hue deve ser UF para o degradê 'magma' funcionar
sns.barplot(
    x='Número de IES Privadas',
    y='UF',
    data=df_private_ies_uf,
    palette='magma',
    hue='UF',
    legend=False
)

plt.title('Distribuição de IES Privadas Ativas por UF')
plt.xlabel('Número de Instituições Privadas')
plt.ylabel('Unidade Federativa (UF)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# um novo DataFrame apenas com os cursos ativos

filtro_ativos = df_cursos['SITUACAO_CURSO'] == 'Em atividade'
df_apenas_ativos = df_cursos[filtro_ativos]

# 2. Contar e exibir os top 50 dessa lista filtrada
print("Top 50 Áreas (Apenas Cursos Ativos):")
display(df_apenas_ativos['AREA_OCDE_CINE'].value_counts().head(50).to_frame(name='Número de Cursos'))

```

```python
# 1. Preparar o Top 50 GERAL (sem filtro)
top50_geral = df_cursos['AREA_OCDE_CINE'].value_counts().head(50).index.tolist()

# 2. Preparar o Top 50 APENAS ATIVOS

df_ativos = df_cursos[df_cursos['SITUACAO_CURSO'] == 'Em atividade']
top50_ativos = df_ativos['AREA_OCDE_CINE'].value_counts().head(50).index.tolist()

# 3. Comparação usando Conjuntos (Sets)
set_geral = set(top50_geral)
set_ativos = set(top50_ativos)

quem_saiu = set_geral - set_ativos   # Estava no Geral, mas NÃO está no Ativo
quem_entrou = set_ativos - set_geral # Está no Ativo, mas NÃO estava no Geral

# 4. Exibir os Resultados
print("=== ANÁLISE DE MUDANÇAS NO RANKING (TOP 50) ===\n")

print(f"🔻 QUEM SAIU ({len(quem_saiu)} áreas):")
print("Essas áreas estavam no Top 50 Geral, mas perderam posição ao filtrar por ativos:")
if not quem_saiu:
    print("  - Nenhuma mudança (as listas são iguais ou as que saíram não afetaram o top 50).")
else:
    for area in quem_saiu:
        print(f"  - {area}")

print("-" * 50)

print(f"🔺 QUEM ENTROU ({len(quem_entrou)} áreas):")
print("Essas áreas ganharam destaque e entraram no Top 50 apenas nos ativos:")
if not quem_entrou:
    print("  - Nenhuma mudança.")
else:
    for area in quem_entrou:
        print(f"  - {area}")
```

```python
import pandas as pd

# 1. CLASSIFICAR IES (No DataFrame Completo)
# Aplica a lógica em todo o histórico primeiro para ter o 'Geral'
def classificar_categoria(cat):
    if 'Pública' in str(cat) or 'Federal' in str(cat) or 'Estadual' in str(cat) or 'Municipal' in str(cat):
        return 'Pública'
    else:
        return 'Privada'

# Cria a coluna auxiliar no DataFrame principal
df_cursos['Tipo_Instituicao'] = df_cursos['CATEGORIA_ADMINISTRATIVA'].apply(classificar_categoria)

# 2. CALCULAR TOTAIS (Histórico Geral)
# Agrupa por Tipo e Modalidade e conta tudo
resumo_total = df_cursos.groupby(['Tipo_Instituicao', 'MODALIDADE']).size().rename('Total Histórico')

# 3. CALCULAR ATIVOS (Apenas em atividade)
# Filtra primeiro, depois agrupa e conta
resumo_ativos = df_cursos[df_cursos['SITUACAO_CURSO'] == 'Em atividade'].groupby(['Tipo_Instituicao', 'MODALIDADE']).size().rename('Ativos Hoje')

# 4. JUNTAR AS INFORMAÇÕES
# Unir as duas contagens numa tabela só
df_comparacao = pd.concat([resumo_total, resumo_ativos], axis=1)

# Se houver algum grupo sem cursos ativos (NaN), preenchemos com 0
df_comparacao = df_comparacao.fillna(0)

# 5. CALCULA A DIFERENÇA E PORCENTAGEM (A "Evolução")
df_comparacao['Extintos/Inativos'] = df_comparacao['Total Histórico'] - df_comparacao['Ativos Hoje']
df_comparacao['% Sobrevivência'] = (df_comparacao['Ativos Hoje'] / df_comparacao['Total Histórico']) * 100

# Ajuste de formatação para ficar bonito
pd.options.display.float_format = '{:.1f}'.format

print("=== COMPARAÇÃO: TOTAL HISTÓRICO vs. SITUAÇÃO ATIVA ===")
display(df_comparacao)
```
